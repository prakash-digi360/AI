```python
# -*- coding: utf-8 -*-
"""
Ethinylestradiol (EE) Drug Development Pipeline Analysis
- Drug Discovery
- Preclinical Development
- Clinical Development
- Regulatory Approval
- Post-Marketing & Pharmacovigilance
- Challenges in Drug Development
- Includes Duration and Cost Estimates for Each Phase
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw, Descriptors, Lipinski, AllChem, DataStructs
from rdkit.Chem.Draw import IPythonConsole
from sklearn.linear_model import LinearRegression

# Phase durations (years) and costs (million USD)
pipeline_data = {
    'Phase': ['Drug Discovery', 'Preclinical Development', 'Clinical Development (Phase 1)',
              'Clinical Development (Phase 2)', 'Clinical Development (Phase 3)',
              'Regulatory Approval', 'Post-Marketing & Pharmacovigilance'],
    'Duration_Years': [3.5, 1.5, 1.5, 2.5, 4.0, 1.0, 7.5],  # Average durations
    'Cost_Million_USD': [300, 100, 35, 75, 200, 30, 125]  # Average costs
}
df_pipeline = pd.DataFrame(pipeline_data)

# =============================================
# 1. DRUG DISCOVERY
# =============================================
# 1.1 Molecular Analysis
ee_smiles = "CC#C[C@]12CC[C@@H]3c4ccc(cc4CC[C@H]3[C@@H]1CC[C@@H]2O)O"
ee_mol = Chem.MolFromSmiles(ee_smiles)

# Generate 2D/3D coordinates
AllChem.Compute2DCoords(ee_mol)
AllChem.EmbedMolecule(ee_mol)

# Calculate properties
mw = Descriptors.MolWt(ee_mol)
logp = Descriptors.MolLogP(ee_mol)
hbd = Lipinski.NumHDonors(ee_mol)
hba = Lipinski.NumHAcceptors(ee_mol)

print(f"""
Drug Discovery - Molecular Properties:
- Duration: {df_pipeline.loc[0, 'Duration_Years']} years
- Estimated Cost: ${df_pipeline.loc[0, 'Cost_Million_USD']} million
- Molecular Weight: {mw:.2f} g/mol
- LogP: {logp:.2f}
- H-bond Donors: {hbd}
- H-bond Acceptors: {hba}
""")

# Plot molecule
img = Draw.MolsToGridImage([ee_mol], molsPerRow=1, subImgSize=(300,300))
display(img)

# 1.2 Target Identification & Validation
target_data = {
    'Compound': ['EE', 'Estradiol', 'Estetrol', 'Control'],
    'Kd_nM': [0.5, 1.0, 2.5, 100]
}
df_target = pd.DataFrame(target_data)

# 1.3 Hit Generation
# High-Throughput Screening (HTS)
hts_data = {
    'Library': ['Small_Molecules', 'Natural_Products', 'Fragments'],
    'Compounds_Screened': [10000, 5000, 2000],
    'Hits': [150, 80, 30]
}
df_hts = pd.DataFrame(hts_data)
df_hts['Hit_Rate'] = df_hts['Hits'] / df_hts['Compounds_Screened'] * 100

# Structure-Based Drug Design (SBDD)
docking_data = {
    'Compound': ['EE', 'Estradiol', 'Estetrol'],
    'Docking_Score': [-8.5, -7.8, -7.2]
}
df_docking = pd.DataFrame(docking_data)

# Fragment-Based Screening
fragment_data = {
    'Fragment': ['F1', 'F2', 'F3'],
    'Binding_Affinity_uM': [50, 20, 30]
}
df_fragment = pd.DataFrame(fragment_data)

# Drug Repurposing
repurpose_data = {
    'Drug': ['Tamoxifen', 'Raloxifene', 'Placebo'],
    'Activity_nM': [10, 15, 1000]
}
df_repurpose = pd.DataFrame(repurpose_data)

# 1.4 Lead Optimization
optimization_data = {
    'Iteration': ['Hit', 'Lead1', 'Lead2', 'Lead3'],
    'LogP': [3.5, 3.2, 3.0, 2.8],
    'Molecular_Weight': [296.4, 290.0, 285.0, 280.0],
    'Binding_Affinity_Kd_nM': [2.0, 1.5, 1.0, 0.8],
    'Caco2_Permeability': [1.0, 1.1, 1.2, 1.3]
}
df_optimization = pd.DataFrame(optimization_data)

# 1.5 Molecular Similarity Comparison
estrogens = {
    'Ethinylestradiol': ee_smiles,
    'Estradiol': 'CC(=O)[C@]12CC[C@H]3[C@H]([C@@H]1CC[C@@H]2O)CCC4=CC(=O)CCC34',
    'Estetrol': 'C[C@]12CC[C@H]3[C@H]([C@@H]1CC[C@@H]2O)CCC4=CC(=O)CCC34O'
}
fps = []
names = []
for name, smi in estrogens.items():
    mol = Chem.MolFromSmiles(smi)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2)
    fps.append(fp)
    names.append(name)

# Similarity matrix
sim_matrix = np.zeros((len(fps), len(fps)))
for i in range(len(fps)):
    for j in range(len(fps)):
        sim_matrix[i,j] = DataStructs.TanimotoSimilarity(fps[i], fps[j])

print(f"""
Drug Discovery Results:
- Target Validation (Estrogen Receptor Kd, nM): {df_target.set_index('Compound')['Kd_nM'].to_dict()}
- HTS Hit Rates (%): {df_hts.set_index('Library')['Hit_Rate'].to_dict()}
- SBDD Docking Scores (kcal/mol): {df_docking.set_index('Compound')['Docking_Score'].to_dict()}
- Fragment Binding Affinities (µM): {df_fragment.set_index('Fragment')['Binding_Affinity_uM'].to_dict()}
- Repurposing Activity (EC50, nM): {df_repurpose.set_index('Drug')['Activity_nM'].to_dict()}
- Lead Optimization (Lead3 vs Hit):
  - LogP: {df_optimization.loc[3, 'LogP']:.2f} (from {df_optimization.loc[0, 'LogP']:.2f})
  - Molecular Weight: {df_optimization.loc[3, 'Molecular_Weight']:.2f} (from {df_optimization.loc[0, 'Molecular_Weight']:.2f})
  - Binding Affinity (Kd): {df_optimization.loc[3, 'Binding_Affinity_Kd_nM']:.2f} nM (from {df_optimization.loc[0, 'Binding_Affinity_Kd_nM']:.2f} nM)
  - Caco2 Permeability: {df_optimization.loc[3, 'Caco2_Permeability']:.2f} nm/s (from {df_optimization.loc[0, 'Caco2_Permeability']:.2f} nm/s)
""")

# Plot drug discovery data
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(24, 5))
ax1.bar(df_target['Compound'], df_target['Kd_nM'], color='purple', alpha=0.7)
ax1.set_title('Estrogen Receptor Binding Affinity', fontsize=14)
ax1.set_xlabel('Compound', fontsize=12)
ax1.set_ylabel('Kd (nM)', fontsize=12)
ax1.grid(alpha=0.3)

ax2.bar(df_hts['Library'], df_hts['Hit_Rate'], color='orange', alpha=0.7)
ax2.set_title('HTS Hit Rates', fontsize=14)
ax2.set_xlabel('Compound Library', fontsize=12)
ax2.set_ylabel('Hit Rate (%)', fontsize=12)
ax2.grid(alpha=0.3)

ax3.plot(df_optimization['Iteration'], df_optimization['Binding_Affinity_Kd_nM'], 'b-', marker='o', label='Binding Affinity (Kd, nM)')
ax3.plot(df_optimization['Iteration'], df_optimization['LogP'], 'r-', marker='s', label='LogP')
ax3.set_title('Lead Optimization Trends', fontsize=14)
ax3.set_xlabel('Optimization Stage', fontsize=12)
ax3.set_ylabel('Value', fontsize=12)
ax3.legend()
ax3.grid(alpha=0.3)

ax4.matshow(sim_matrix, cmap='viridis')
ax4.set_xticks(range(len(names)))
ax4.set_yticks(range(len(names)))
ax4.set_xticklabels(names, rotation=45)
ax4.set_yticklabels(names)
plt.colorbar(ax4.matshow(sim_matrix, cmap='viridis'), ax=ax4, label='Tanimoto Similarity')
ax4.set_title('Structural Similarity of Estrogens', fontsize=14)
plt.tight_layout()
plt.show()

# =============================================
# 2. PRECLINICAL DEVELOPMENT
# =============================================
# 2.1 In Vitro & In Vivo Testing
# In vitro: ADME Prediction
adme_data = {
    'Compound': ['EE', 'Estradiol', 'Placebo'],
    'Caco2_Permeability': [1.2, 0.8, 0.1],
    'HIA': [85, 92, 5],
    'PPB': [98, 95, 10],
    'CYP3A4_Substrate': [1, 1, 0]
}
df_adme = pd.DataFrame(adme_data).set_index('Compound')

# Plot ADME properties
fig, ax = plt.subplots(2, 2, figsize=(12, 10))
df_adme['Caco2_Permeability'].plot.bar(ax=ax[0,0], title='Caco-2 Permeability (nm/s)')
df_adme['HIA'].plot.bar(ax=ax[0,1], title='Human Intestinal Absorption (%)')
df_adme['PPB'].plot.bar(ax=ax[1,0], title='Plasma Protein Binding (%)')
df_adme['CYP3A4_Substrate'].plot.bar(ax=ax[1,1], title='CYP3A4 Substrate (Yes=1)')
plt.tight_layout()
plt.show()

# In vitro: CYP inhibition (IC50 in µM)
cyp_data = {
    'Enzyme': ['CYP1A2', 'CYP2C9', 'CYP2C19', 'CYP2D6', 'CYP3A4'],
    'IC50_uM': [50, 30, 25, 40, 15]
}
df_cyp = pd.DataFrame(cyp_data)

# In vivo: Rat PK and toxicity
rat_pk_data = {
    'Time_h': [0, 0.5, 1, 2, 4, 8, 12, 24],
    'Concentration_ng_mL': [0, 15, 35, 50, 40, 25, 10, 3]
}
df_rat_pk = pd.DataFrame(rat_pk_data)

toxicity_data = {
    'Parameter': ['LD50', 'NOAEL'],
    'Value_mg_kg': [500, 50]
}
df_toxicity = pd.DataFrame(toxicity_data)

# Bioanalytical Method (LC-MS Analysis)
calibration_data = {
    'Concentration_ng_mL': [0, 0.5, 1, 2, 5, 10, 20, 50, 100],
    'Peak_Area': [0, 500, 950, 2100, 4800, 10500, 19800, 49500, 101000]
}
df_cal = pd.DataFrame(calibration_data)

# Linear regression for calibration curve
X = df_cal['Concentration_ng_mL'].values.reshape(-1, 1)
y = df_cal['Peak_Area'].values
reg = LinearRegression().fit(X, y)
slope = reg.coef_[0]
intercept = reg.intercept_
r_squared = reg.score(X, y)

# Human PK data
pk_data = {
    'Time_h': [0, 0.5, 1, 2, 4, 6, 8, 12, 24],
    'Concentration_ng_mL': [0, 10, 25, 40, 30, 20, 15, 8, 2]
}
df_pk = pd.DataFrame(pk_data)

# Calculate PK parameters
rat_auc = np.trapz(df_rat_pk['Concentration_ng_mL'], df_rat_pk['Time_h'])
rat_c_max = df_rat_pk['Concentration_ng_mL'].max()
human_auc = np.trapz(df_pk['Concentration_ng_mL'], df_pk['Time_h'])
human_c_max = df_pk['Concentration_ng_mL'].max()

# 2.2 IND Application
ind_data = {
    'Parameter': ['Efficacy', 'Safety', 'PK'],
    'Metric': ['EC50 = 0.5 nM', 'NOAEL = 50 mg/kg', f'AUC = {rat_auc:.2f} ng·h/mL']
}
df_ind = pd.DataFrame(ind_data)

print(f"""
Preclinical Development Results:
- Duration: {df_pipeline.loc[1, 'Duration_Years']} years
- Estimated Cost: ${df_pipeline.loc[1, 'Cost_Million_USD']} million
- ADME Properties: {df_adme.to_dict()}
- CYP Inhibition (IC50, µM): {df_cyp.set_index('Enzyme')['IC50_uM'].to_dict()}
- Rat PK:
  - AUC (0-24h): {rat_auc:.2f} ng·h/mL
  - C_max: {rat_c_max:.2f} ng/mL
- Human PK:
  - AUC (0-24h): {human_auc:.2f} ng·h/mL
  - C_max: {human_c_max:.2f} ng/mL
- LC-MS Calibration:
  - Slope: {slope:.2f}
  - Intercept: {intercept:.2f}
  - R-squared: {r_squared:.3f}
- Toxicity:
  - LD50 (Rat, Oral): {df_toxicity.loc[0, 'Value_mg_kg']:.2f} mg/kg
  - NOAEL (Rat, Oral): {df_toxicity.loc[1, 'Value_mg_kg']:.2f} mg/kg
- IND Application Summary: {df_ind.set_index('Parameter')['Metric'].to_dict()}
""")

# Plot preclinical data
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(18, 5))
ax1.bar(df_cyp['Enzyme'], df_cyp['IC50_uM'], color='green', alpha=0.7)
ax1.set_title('CYP Inhibition IC50', fontsize=14)
ax1.set_xlabel('Enzyme', fontsize=12)
ax1.set_ylabel('IC50 (µM)', fontsize=12)
ax1.grid(alpha=0.3)

ax2.plot(df_rat_pk['Time_h'], df_rat_pk['Concentration_ng_mL'], 'b-', marker='o', label='Rat')
ax2.plot(df_pk['Time_h'], df_pk['Concentration_ng_mL'], 'r-', marker='s', label='Human')
ax2.set_title('PK Profiles', fontsize=14)
ax2.set_xlabel('Time (h)', fontsize=12)
ax2.set_ylabel('Concentration (ng/mL)', fontsize=12)
ax2.legend()
ax2.grid(alpha=0.3)

ax3.scatter(df_cal['Concentration_ng_mL'], df_cal['Peak_Area'], color='blue', label='Data')
ax3.plot(df_cal['Concentration_ng_mL'], reg.predict(X), 'r-', label=f'Fit (R²={r_squared:.3f})')
ax3.set_title('LC-MS Calibration Curve', fontsize=14)
ax3.set_xlabel('Concentration (ng/mL)', fontsize=12)
ax3.set_ylabel('Peak Area', fontsize=12)
ax3.legend()
ax3.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# =============================================
# 3. CLINICAL DEVELOPMENT (HUMAN TRIALS)
# =============================================
# Phase 1: Safety & Tolerability
phase1_data = {
    'Dose_ug': [10, 20, 30, 50],
    'Adverse_Events': [2, 5, 8, 15],
    'C_max_ng_mL': [5, 10, 15, 25]
}
df_phase1 = pd.DataFrame(phase1_data)

# Phase 2: Efficacy & Side Effects
phase2_data = {
    'Dose_ug': [20, 30, 40],
    'Efficacy_Percent': [60, 75, 85],
    'Side_Effects': [10, 15, 20]
}
df_phase2 = pd.DataFrame(phase2_data)

# Phase 3: Large-Scale Testing
phase3_data = {
    'Year': [2018, 2019, 2020, 2021],
    'Patients': [500, 1000, 1500, 2000],
    'Success_Rate': [70, 72, 75, 78]
}
df_phase3 = pd.DataFrame(phase3_data)

# Phase 4: Post-Marketing Surveillance
phase4_data = {
    'Year': [2022, 2023, 2024, 2025],
    'Adverse_Events_Per_1000': [5, 4, 3, 2]
}
df_phase4 = pd.DataFrame(phase4_data)

# Historical Clinical Trial Data (Dose & VTE Risk)
years = np.arange(1960, 2023)
ee_doses = np.clip(np.exp(-0.03*(years-1960))*150, 20, 150)  # Exponential dose reduction
vte_risk = np.clip(0.1 + 0.0003*(years-1960)**2, 0.1, 3.5)  # Quadratic risk increase

print(f"""
Clinical Development Results:
- Phase 1 (Safety):
  - Duration: {df_pipeline.loc[2, 'Duration_Years']} years
  - Estimated Cost: ${df_pipeline.loc[2, 'Cost_Million_USD']} million
  - Max Dose Tested: {df_phase1['Dose_ug'].max()} µg
  - Max Adverse Events: {df_phase1['Adverse_Events'].max()}
  - Max C_max: {df_phase1['C_max_ng_mL'].max()} ng/mL
- Phase 2 (Efficacy):
  - Duration: {df_pipeline.loc[3, 'Duration_Years']} years
  - Estimated Cost: ${df_pipeline.loc[3, 'Cost_Million_USD']} million
  - Max Efficacy: {df_phase2['Efficacy_Percent'].max()}%
  - Max Side Effects: {df_phase2['Side_Effects'].max()}%
- Phase 3 (Large-Scale):
  - Duration: {df_pipeline.loc[4, 'Duration_Years']} years
  - Estimated Cost: ${df_pipeline.loc[4, 'Cost_Million_USD']} million
  - Max Success Rate: {df_phase3['Success_Rate'].max()}% in {df_phase3['Year'].iloc[-1]}
- Phase 4 (Post-Marketing):
  - Duration: {df_pipeline.loc[6, 'Duration_Years']} years (ongoing)
  - Estimated Cost: ${df_pipeline.loc[6, 'Cost_Million_USD']} million
  - Latest Adverse Events: {df_phase4['Adverse_Events_Per_1000'].iloc[-1]} per 1000 patients
""")

# Plot clinical data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(df_phase1['Dose_ug'], df_phase1['Adverse_Events'], 'r-', marker='o', label='Adverse Events')
ax1.plot(df_phase1['Dose_ug'], df_phase1['C_max_ng_mL'], 'b-', marker='s', label='C_max (ng/mL)')
ax1.set_title('Phase 1: Safety & PK', fontsize=14)
ax1.set_xlabel('Dose (µg)', fontsize=12)
ax1.set_ylabel('Value', fontsize=12)
ax1.legend()
ax1.grid(alpha=0.3)

ax2.plot(df_phase3['Year'], df_phase3['Success_Rate'], 'g-', marker='o')
ax2.set_title('Phase 3: Success Rate Over Time', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('Success Rate (%)', fontsize=12)
ax2.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Plot historical clinical trial data
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.plot(years, ee_doses, 'r-', linewidth=2)
ax1.set_title('EE Dose Reduction Over Time', fontsize=14)
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Dose (µg)', fontsize=12)
ax1.grid(alpha=0.3)

ax2.plot(years, vte_risk, 'b--', linewidth=2)
ax2.set_title('Reported VTE Risk per 10,000 Women', fontsize=14)
ax2.set_xlabel('Year', fontsize=12)
ax2.set_ylabel('VTE Cases', fontsize=12)
ax2.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# =============================================
# 4. REGULATORY APPROVAL
# =============================================
regulatory_data = {
    'Metric': ['Efficacy', 'Safety', 'PK Consistency'],
    'Value': ['78% Success Rate', 'Max 15% Side Effects', 'Stable AUC/C_max']
}
df_regulatory = pd.DataFrame(regulatory_data)

print(f"""
Regulatory Approval Summary:
- Duration: {df_pipeline.loc[5, 'Duration_Years']} years
- Estimated Cost: ${df_pipeline.loc[5, 'Cost_Million_USD']} million
- Metrics: {df_regulatory.set_index('Metric')['Value'].to_dict()}
""")

# =============================================
# 5. POST-MARKETING & PHARMACOVIGILANCE
# =============================================
print(f"""
Post-Marketing & Pharmacovigilance:
- Duration: {df_pipeline.loc[6, 'Duration_Years']} years (ongoing)
- Estimated Cost: ${df_pipeline.loc[6, 'Cost_Million_USD']} million
- Adverse Event Trend (per 1000 patients): {df_phase4.set_index('Year')['Adverse_Events_Per_1000'].to_dict()}
""")

# Plot pharmacovigilance data
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(df_phase4['Year'], df_phase4['Adverse_Events_Per_1000'], 'b-', marker='o')
ax.set_title('Post-Marketing Adverse Events', fontsize=14)
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Adverse Events per 1000 Patients', fontsize=12)
ax.grid(alpha=0.3)
plt.show()

# =============================================
# 6. CHALLENGES IN DRUG DEVELOPMENT
# =============================================
total_duration = df_pipeline['Duration_Years'].sum()
total_cost = df_pipeline['Cost_Million_USD'].sum()
print(f"""
Challenges in Drug Development:
- High Costs: Total estimated cost ~${total_cost} million, with significant expenses in Drug Discovery (${df_pipeline.loc[0, 'Cost_Million_USD']}M) and Clinical Phase 3 (${df_pipeline.loc[4, 'Cost_Million_USD']}M).
- Long Timelines: Total duration ~{total_duration:.1f} years, spanning {df_pipeline.loc[0, 'Duration_Years']} years for discovery to {df_pipeline.loc[4, 'Duration_Years']} years for Phase 3 trials.
- High Failure Rates: ~90% of drug candidates fail during clinical trials due to efficacy or safety issues.
- Target Validation: Identifying and validating the right biological target is complex and risky.
- Regulatory Hurdles: Stringent requirements for safety and efficacy data can delay approval, adding ~{df_pipeline.loc[5, 'Duration_Years']} years.
- Drug-Drug Interactions: Potential interactions (e.g., CYP inhibition) complicate development.
- Patient Variability: Genetic and environmental factors affect drug response, complicating trial design.
- Post-Marketing Risks: Rare adverse events may emerge, requiring ongoing monitoring for ~{df_pipeline.loc[6, 'Duration_Years']} years.
""")

# =============================================
# 7. PIPELINE SUMMARY VISUALIZATION
# =============================================
print(f"""
Pipeline Summary:
{df_pipeline.set_index('Phase')[['Duration_Years', 'Cost_Million_USD']].to_dict()}
""")

# Plot duration and cost
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))
ax1.bar(df_pipeline['Phase'], df_pipeline['Duration_Years'], color='skyblue', alpha=0.7)
ax1.set_title('Pipeline Phase Durations', fontsize=14)
ax1.set_xlabel('Phase', fontsize=12)
ax1.set_ylabel('Duration (Years)', fontsize=12)
ax1.set_xticklabels(df_pipeline['Phase'], rotation=45, ha='right')
ax1.grid(alpha=0.3)

ax2.bar(df_pipeline['Phase'], df_pipeline['Cost_Million_USD'], color='lightcoral', alpha=0.7)
ax2.set_title('Pipeline Phase Costs', fontsize=14)
ax2.set_xlabel('Phase', fontsize=12)
ax2.set_ylabel('Cost (Million USD)', fontsize=12)
ax2.set_xticklabels(df_pipeline['Phase'], rotation=45, ha='right')
ax2.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
