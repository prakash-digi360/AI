```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Hypothetical historical annual revenue data for Semaglutide (USD Billion)
years_hist = np.array([2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024])
revenue_hist = np.array([0.1, 0.3, 0.7, 1.5, 3.2, 6.5, 12.0, 18.5, 24.0, 28.4])

# Fit linear regression model to historical data
slope, intercept, r_value, p_value, std_err = stats.linregress(years_hist, revenue_hist)

print(f"Fitted model: revenue = {slope:.2f} * year + {intercept:.2f}")
print(f"R-squared: {r_value**2:.3f}")

# Forecast for next 5 years
years_forecast = np.arange(2025, 2030)
revenue_forecast = slope * years_forecast + intercept

# Plot
plt.figure(figsize=(10,6))
plt.plot(years_hist, revenue_hist, 'bo-', label='Historical Revenue')
plt.plot(years_forecast, revenue_forecast, 'ro--', label='Forecasted Revenue')
plt.fill_between(years_forecast,
                 revenue_forecast - 1.96 * std_err * np.sqrt(years_forecast - years_forecast[0]),
                 revenue_forecast + 1.96 * std_err * np.sqrt(years_forecast - years_forecast[0]),
                 color='pink', alpha=0.3, label='95% Confidence Interval')

plt.title('Time Series & Forecast Analysis for Semaglutide Revenue')
plt.xlabel('Year')
plt.ylabel('Revenue (USD Billion)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
